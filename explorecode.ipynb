```python
fig, ax1 = plt.subplots(figsize=(9, 6))
x=pd.DatetimeIndex(df_us['Date']).year
y1=df_us['Consumer_Price_Index']
y2=df_us['Mortgage_Rate']

# Instantiate a second axes that shares the same x-axis
ax2 = ax1.twinx()  
ax1.bar(x,y1, color="orange")
ax2.plot(x,y2, color="blue", marker=".", markersize=10)
ax1.set_xlabel("Date")
ax1.set_ylabel("Consumer_Price_Index", color= 'orange')
ax2.set_ylabel("Mortgage_Rate(%)",  color= 'blue')
plt.show()
```

```python
y1.corr(y2)
```

```python
x0=pd.DatetimeIndex(df_us['Date']).year
x1=df_us['Stock_Price_Index']
x2=df_us['Consumer_Price_Index']

plt.subplots(2, 1, sharex=True, figsize=(9, 6))

plt.subplot(2,1,1)
plt.plot(x0,x1)
plt.fill_between(x0,x1,color="red")
plt.title("Stock_Price_Index")

plt.subplot(2,1,2)
plt.plot(x0,x2,marker="^")
plt.title("Consumer_Price_Index")

plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
sns.regplot(data=df_us, x= df_us['House_Price_Index'], y=df_us['Mortgage_Rate'])
```

```python
sns.regplot(data=df_us, x=df_us['Consumer_Price_Index'], y= df_us['Mortgage_Rate'])
```

```python
sns.regplot(data=df_us, x=df_us['Real_Disposable_Income'], y= df_us['Mortgage_Rate'])
```

```python
lm = LinearRegression()
```

```python
indep = df_us[['House_Price_Index','Consumer_Price_Index', 'Real_Disposable_Income']]
dep = df_us['Mortgage_Rate']
```

```python
lm.fit(indep, dep)
```

```python
HIPredect = lm.predict(indep)
```

```python
ax1 = sns.distplot(df_us['Mortgage_Rate'], hist=False, color="r", label="Actual Value")
sns.distplot(HIPredect, hist=False, color="b", label="Fitted Values" , ax=ax1)


plt.title('Actual vs Fitted Values for Mortgage Rate')
plt.xlabel('Mortgage Rate')
plt.ylabel('Rates')

plt.show()
plt.close()
```

```python
lm2 = LinearRegression()
```

```python
lm2.fit(df_us[['Consumer_Price_Index']], df_us['Mortgage_Rate'])
```

```python
newrate = lm2.predict(df_us[['Consumer_Price_Index']])
```

```python
ax1 = sns.scatterplot(df_us['Mortgage_Rate'], color="r", label="Actual Value")
sns.scatterplot(newrate, color="b", label="Fitted Values" , ax=ax1)


plt.title('Actual vs Fitted Values for Mortgage Rate')
plt.xlabel('Mortgage Rate')
plt.ylabel('Rates')

plt.show()
plt.close()
```
